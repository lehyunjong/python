```python
cost = int(input("1:주택용,2:공업용,3:산업용?"))
use = int(input("사용량(kwh)?"))
if cost == 1:
  print(f"용도:{cost}, 사용량:{use:,.2f}, 전기요금:{(use*88)+910:,.2f}")
else :
  if cost == 2:
    print(f"용도:{cost}, 사용량:{use:,.2f}, 전기요금:{(use*182)+1600:,.2f}")
  else :
    if cost == 3:
      print(f"용도:{cost}, 사용량:{use:,.2f}, 전기요금:{(use*275)+7300:,.2f}")


```

```python
name = input("이름 : ")
kor = int(input("국어점수:"))
eng = int(input("영어점수:"))
math = int(input("수학점수:"))
sum = kor+eng+math
age = (kor+eng+math)/3
if age >= 95:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 A+ 학점 입니다.")
elif age >= 90:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 A 학점 입니다.")
elif age >= 85:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 B+ 학점 입니다.")
elif age >= 80:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 B 학점 입니다.")
elif age >= 75:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 C+ 학점 입니다.")
elif age >= 70:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 C 학점 입니다.")
elif age >= 65:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 D+ 학점 입니다.")
elif age >= 60:
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 D 학점 입니다.")
else :
  print(f"{name}의 총점은 {sum}점, 평균은 {age:,.2f}점, 학점은 F 학점 입니다.")

```
