```python
def myFunction():
  langStr=("Java","C","Python","R","Web")
  x = list(langStr)
  x.sort(key = len, reverse = True)
  a = list(map(lambda x:len(x), langStr))
  a.sort(reverse = True)
  for i in range(0,5):
    print(f"{x[i]} \t {a[i]}")


myFunction()
```
