```python
movie_list=[("오징어게임",50.45), ("이터널스",35.46),("그레비티",9.8), ("노타임투다이",52.5),
("스파이더맨",15.47)]
movie_dict = dict.fromkeys(movie_list)
x = sorted(movie_dict , key = lambda x:x[1] , reverse = True)
y = 1

print("영화제목    예매율    순위")
print("========    ======    ====")
for k, v in x:
  print(k , "{0: >5}".format(v) ,f"     {y}")

  y += 1


```
