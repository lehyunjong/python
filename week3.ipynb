```python
A = int(input("소프트웨어 설계 : "))
B = int(input("소프트웨어 개발 : "))
C = int(input("데이터베이스 구축 : "))
D = int(input("프로그래밍언어 활용 : "))
E = int(input("정보시스템 구축 관리 : "))
avg = (A+B+C+D+E)/5
print("평균 :",int(avg))
if A >= 40 and B >= 40 and C >= 40 and D >= 40 and E >= 40 and avg >= 60:
  print("결과 : 합격")
else:
  print("결과 : 불합격")
```

```python
number = int(input("어떤 수?"))
if number > 100:
  print(f"입력한 수 {number}은(는) 100보다 큽니다")
else:
  if number > 50:
    print(f"입력한 수 {number}은(는) 50보다 크고 100보다 작습니다.")
  else:
    print(f"입력한 수 {number}은(는) 50보다 작습니다.")
```

```python
number = int(input("어떤 수?"))
if 0 <= number < 10:
  print("한 자릿수 숫자입니다.")
elif 10 <= number < 100:
  print("두 자릿수 숫자입니다.") 
elif 100 <= number < 1000:
  print("세 자릿수 숫자입니다.") 
elif 1000 <= number < 10000:
  print("네 자릿수 숫자입니다.") 
```
