```python
for i in range(1,6,1) :
  for j in range(i):
    print(" ", end="")
  for j in range(6-i,0,-1) :
    print(i, end="")
    i+=1
  print()
```

```python
x = int(input("숫자 입력:"))
y = 0
for i in range(2,x):
  if x % i == 0:
    y=1

if y == 0:
  print(f"{x}은(는) 소수 입니다.")
else:
  print(f"{x}은(는) 소수가 아닙니다.")
```
